# Criando uma rede
Esse documento demonstra, passo a passo, a criação de uma rede para uma simulação.

```python
# Importando as dependências
from QKDnet.components import Network, Controller
```

Como o código do repositório não é uma biblioteca do Python (pelo menos ainda não), é mais fácil criar os scripts aqui na raiz. Ou seja, para importar as dependências da mesma forma que é feito acima, os novos scripts devem ser irmãos deste documento.

```python
# Instanciando a rede
rede = Network()
# Definindo as características da rede
rede.set_topology("Lattice", 3, 4)
rede.set_capacity(1)
rede.set_fidelity(1)
rede.set_nqubits(100)
rede.draw()
```

Ao instanciar a rede, os valores para capacidade, fidelidade e número de qubits já serão definidos (com os valores 1, 1 e 100, respectivamente). Mas podem ser alterados com os métodos acima. Isso inclui também a topologia.

```python
# Instanciando o controlador
controlador = Controller(rede)
controlador.set_sorter("urgency")
controlador.set_path_finder("kshortest")
```

O controlador também pode ser somente instanciado sem a necessidade de especificar seu ``sorter`` e ``path_finder``. Suas configurações padrões são as utilizadas no código de exemplo. Para saber as outras possibilidades e o que fazem, basta consultar a docstring da função, ou seus arquivos.
