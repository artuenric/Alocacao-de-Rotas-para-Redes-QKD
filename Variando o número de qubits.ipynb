```python
# --- 1. Importações e Configuração do Ambiente ---
import sys
import os
import matplotlib.pyplot as plt

# Adiciona o diretório do projeto ao path para encontrar o módulo QKDnet
# Certifique-se de que este notebook está na pasta raiz do projeto
sys.path.insert(0, os.path.abspath('.'))

from QKDnet.components import Network, Controller
from QKDnet.sim import Simulation
from QKDnet.utils import Logger, float_to_string

# --- 2. Parâmetros da Simulação ---
n_qubits_lista = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 2500]
n_requests = 1000
capacity = 10
apps_distribution = [0.33, 0.33, 0.33]
simulation_case = 3
estrategia = 'urgency'

# Lista para armazenar os resultados para o gráfico
requisicoes_atendidas = []

# --- 3. Execução das Simulações ---
print(f'--- Iniciando simulações para a estratégia: {estrategia.upper()} ---')

for q in n_qubits_lista:
    # Rede
    rede = Network()
    rede.set_topology("Vienna")
    rede.set_nqubits(q)  # Alterando o número de qubits
    rede.set_capacity(capacity)
    
    # Controlador
    controlador = Controller(rede)
    controlador.set_sorter(estrategia)
    
    # Simulação
    sim = Simulation(rede, controlador)
    sim.set_case(simulation_case)
    sim.set_apps_distribution(apps_distribution)
    sim.set_n_requests(n_requests)
    
    # Rodando a simulação
    sim.run()
    
    # Coletando e armazenando os dados
    num_atendidas = len(sim.controller.data_base.served_requests)
    requisicoes_atendidas.append(num_atendidas)
    
    print(f'Nº de Qubits: {q}, Requisições Atendidas: {num_atendidas}')

print('\n--- Simulações concluídas. Gerando o arquivo PDF... ---')

# --- 4. Plotagem e Exportação do Gráfico em PDF ---
plt.style.use('seaborn-v0_8-paper')

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(n_qubits_lista, requisicoes_atendidas, color='tab:blue', marker='o', linestyle='-')
#ax.set_title(f'Requisições Atendidas vs. Número de Qubits (Estratégia: {estrategia.capitalize()})')
ax.set_xlabel('Número de Qubits')
ax.set_ylabel('Nº de Requisições Atendidas')
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
fig.tight_layout()

# Salva a figura em formato PDF
nome_arquivo = "grafico_requisicoes_atendidas_vs_qubits_vienna.pdf"
plt.savefig(nome_arquivo, format='pdf')
print(f"Gráfico salvo como '{nome_arquivo}'")

# Exibe o gráfico no notebook
plt.show()
```
