```python
import networkx as nx
import matplotlib.pyplot as plt
def create_china_topology(seed):
# Create an empty graph
    G = nx.Graph()

    # Add nodes
    for i in range(1, 41):
        G.add_node(i, color='red' if i <= 13 else 'green')

    # Add TR nodes
    for i in range(1, 4):
        G.add_node(f'TR-{i}', color='blue')

    # Add OS nodes
    for i in range(1, 4):
        G.add_node(f'OS-{i}', color='purple')

    # Connect nodes in stars
    for i in range(1, 6):
        G.add_edge(f'OS-1', i)

    for i in range(6, 8):
        G.add_edge(f'OS-2', i)

    for i in range(8, 14):
        G.add_edge(f'OS-3', i)

    for i in range(14, 27):
        G.add_edge(f'TR-1', i)

    for i in range(27, 33):
        G.add_edge(f'TR-2', i)

    for i in range(33, 41):
        G.add_edge(f'TR-3', i)

    # Add direct connections
    G.add_edge('OS-1', 'TR-2')
    G.add_edge('OS-2', 'TR-2')
    G.add_edge('OS-3', 'TR-3')
    G.add_edge('TR-3', 'TR-2')
    G.add_edge('TR-2', 'TR-1')
    G.add_edge('TR-3', 'TR-1')
   
    return G

# Draw the graph
G = create_china_topology(seed=42)
pos = nx.spring_layout(G)
node_colors = [G.nodes[n]['color'] for n in G.nodes]
nx.draw(G, pos, node_color=node_colors, with_labels=True, font_weight='bold')
plt.show()

```
