# Simulação simples
Este documento demonstra a criação básica de uma simulação simples.

```python
# Importando as dependências
from QKDnet.components import Network, Controller
from QKDnet.sim import Simulation
from QKDnet.utils import Logger

# Ativar os logs:
Logger.activate(Logger)
```

Para as simulações, na maioria das vezes, não será necessário mais do que estes ``import``.

```python
# Rede
rede = Network()
rede.set_topology("Lattice", 3,4)
rede.set_nqubits(1000)
rede.draw()
# Controlador
controlador = Controller(rede)
```

Para mais detalhes sobre a rede e o controlador, veja o arquivo ``Rede e Controlador``.

```python
# Simulação
sim = Simulation(rede, controlador)
sim.set_n_requests(10)
sim.set_case(1)
sim.set_apps_distribution([0.33, 0.33, 0.33])
sim.set_max_time_request(10)
sim.run()
```

Os métodos acima alteram respectivamente as propriedades:
- ``n_requests``: o número de requests que o controlador receberá por vez.
- ``case``: os casos variam de 1 a 9 e representam a distribuição para as classes (tamanho da chave necessária) na geração da lista de requests.
- ``apps_distribution``: a distribuição para as apps disponíveis, ou seja, BB84, E91 e B92. Essa ordem deve ser levada em consideração, então atenção.

```python
print(sim.throughput)
```
