```python
# --- 1. Importações e Configuração do Ambiente ---
import matplotlib.pyplot as plt
from QKDnet.components import Network, Controller
from QKDnet.sim import Simulation
from QKDnet.utils import Logger

# --- 2. Parâmetros da Simulação ---
n_requests = 100
n_qubits = 1000
capabilities = [1, 2, 3, 4, 5, 10, 15, 20, 25, 50, 100]
apps_distribution = [0.33, 0.33, 0.33]
simulation_case = 8
estrategia = 'fifo'

# Listas para armazenar os resultados para o gráfico
requisicoes_atendidas = []
tempos_finais = []

# --- 3. Execução das Simulações ---
print(f'--- Iniciando simulações para a estratégia: {estrategia.upper()} ---')

for c in capabilities:
    rede = Network()
    rede.set_topology("Vienna")
    rede.set_nqubits(n_qubits)
    rede.set_capacity(c)
    
    controlador = Controller(rede)
    controlador.set_sorter(estrategia)
    
    sim = Simulation(rede, controlador)
    sim.set_case(simulation_case)
    sim.set_apps_distribution(apps_distribution)
    sim.set_n_requests(n_requests)
    
    sim.run()
    
    num_atendidas = len(sim.controller.data_base.served_requests)
    tempo_final = sim.controller.data_base.final_time
    
    requisicoes_atendidas.append(num_atendidas)
    tempos_finais.append(tempo_final)
    
    print(f'Capacidade: {c}, Requisições Atendidas: {num_atendidas}, Tempo Final: {tempo_final}')

print('\n--- Simulações concluídas. Gerando os arquivos PDF... ---')

# --- 4. Plotagem e Exportação em PDFs Separados ---
plt.style.use('seaborn-v0_8-paper')

# --- Gráfico 1: Requisições Atendidas ---
fig1, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(capabilities, requisicoes_atendidas, color='tab:blue', marker='o', linestyle='-')
#ax1.set_title(f'Requisições Atendidas vs. Capacidade do Canal (Estratégia: {estrategia.capitalize()})')
ax1.set_xlabel('Capacidade do Canal')
ax1.set_ylabel('Nº de Requisições Atendidas')
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
fig1.tight_layout()
nome_arquivo1 = "grafico_requisicoes_atendidas_vs_capacidade_vienna.pdf"
plt.savefig(nome_arquivo1, format='pdf')
print(f"Gráfico 1 salvo como '{nome_arquivo1}'")

# --- Gráfico 2: Tempo Final ---
fig2, ax2 = plt.subplots(figsize=(10, 6))
ax2.plot(capabilities, tempos_finais, color='tab:red', marker='s', linestyle='--')
#ax2.set_title(f'Tempo Final da Simulação vs. Capacidade do Canal (Estratégia: {estrategia.capitalize()})')
ax2.set_xlabel('Capacidade do Canal')
ax2.set_ylabel('Tempo Final da Simulação')
ax2.grid(True, which='both', linestyle='--', linewidth=0.5)
fig2.tight_layout()
nome_arquivo2 = "grafico_tempo_final_vs_capacidade_vienna.pdf"
plt.savefig(nome_arquivo2, format='pdf')
print(f"Gráfico 2 salvo como '{nome_arquivo2}'")

# Exibe os gráficos no notebook
plt.show()
```
