## Rede e Controlador

A Network é a classe responsável pela rede e a classe Controller, pelo controlador. Esses são os objetos básicos de toda simulação. 

```python
# Definindo rede e controlador
rede = Network()
rede.set_topology("China")
controlador = Controller(rede)

# Algumas propriedades da rede
rede.nqubits = 300
rede.neprs = 3
rede.set_fidelity(0.95)
rede.draw()
```

```python
# Calculando a rota mais curta entre dois nós
route = controlador.calculate_shortest_route(1, 5) # Antigo
route = controlador.pathFinder.get_paths(1, 5) # Novo
print(route)
```
